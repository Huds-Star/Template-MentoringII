#### b-) Ler o dataset "*Hourly_wages/hourly_wages.csv*" como dataframe utilizando a biblioteca Pandas e mostrar as 5 primeiras e as 5 últimas linhas do dataset.

```
import pandas as pd

file_path = 'DSBD/Datasets/Hourly_wages/hourly_wages.csv'
df = pd.read_csv(file_path)

print("As 5 primeiras linhas do dataset:")
print(df.head())

print("\nAs 5 últimas linhas do dataset:")
print(df.tail())
```

#### c-) Obter um resumo da estatística descritiva dos dados utilizando um único comando (função).

```
print(df.describe())
```

#### d-) Calcule a proporção de funcionários de sexo masculino e feminino.

```
gender_proportion = df['female'].value_counts(normalize=True)
print("Proporção de funcionários por sexo:")
print(f"Masculino: {gender_proportion[0]:.2%}")
print(f"Feminino: {gender_proportion[1]:.2%}")
```

#### e-) Calcule a proporção de casados e solteiros.

```
marr_proportion = df['marr'].value_counts(normalize=True)
print("Proporção de funcionários por estado civil:")
print(f"Solteiro: {marr_proportion[0]:.2%}")
print(f"Casado: {marr_proportion[1]:.2%}")
```

#### f-) Calcule as médias de cada uma das variáveis: anos de educação, anos de experiência e idade

```
mean_education = df['education_yrs'].mean()
mean_experience = df['experience_yrs'].mean()
mean_age = df['age'].mean()

print(f"Média de anos de educação: {mean_education:.2f}")
print(f"Média de anos de experiência: {mean_experience:.2f}")
print(f"Média de idade: {mean_age:.2f}")
```

#### g-) Calcule as medianas de cada uma das variáveis: anos de educação, anos de experiência e idade.

```
median_education = df['education_yrs'].median()
median_experience = df['experience_yrs'].median()
median_age = df['age'].median()

print(f"Mediana de anos de educação: {median_education:.2f}")
print(f"Mediana de anos de experiência: {median_experience:.2f}")
print(f"Mediana de idade: {median_age:.2f}")
```

####h-) Calcule as modas de cada uma das variáveis: anos de educação, anos de experiência e idade.

```
mode_education = df['education_yrs'].mode()[0] # mode can return multiple values
mode_experience = df['experience_yrs'].mode()[0]
mode_age = df['age'].mode()[0]

print(f"Moda de anos de educação: {mode_education}")
print(f"Moda de anos de experiência: {mode_experience}")
print(f"Moda de idade: {mode_age}")
```

####i-) Analisando especificamente a variável **idade**, responda: qual é a idade do conjunto de dados em que **70%** dos funcionários apresentam idade inferior?

```
age_70th_percentile = df['age'].quantile(0.70)
print(f"A idade em que 70% dos funcionários apresentam idade inferior é: {age_70th_percentile:.2f} anos")
```

####j-) Analisando especificamente a variável **anos de experiência**, responda: qual é a quantidade de anos de experiência no conjunto de dados em que 10% dos funcionários ainda não possuem?

```
experience_10th_percentile = df['experience_yrs'].quantile(0.10)
print(f"A quantidade de anos de experiência em que 10% dos funcionários ainda não possuem é: {experience_10th_percentile:.2f} anos")
```

####k-) Calcule o desvio-padrão de cada um dos atributos: anos de educação, anos de experiência e idade.

```
std_education = df['education_yrs'].std()
std_experience = df['experience_yrs'].std()
std_age = df['age'].std()

print(f"Desvio-padrão de anos de educação: {std_education:.2f}")
print(f"Desvio-padrão de anos de experiência: {std_experience:.2f}")
print(f"Desvio-padrão de idade: {std_age:.2f}")
```

####l-) Determine qual dos três atributos apresenta maior variabilidade: anos de educação, anos de experiência ou idade? Justifique a sua resposta.

```
std_education = df['education_yrs'].std()
std_experience = df['experience_yrs'].std()
std_age = df['age'].std()

print(f"Desvio-padrão de anos de educação: {std_education:.2f}")
print(f"Desvio-padrão de anos de experiência: {std_experience:.2f}")
print(f"Desvio-padrão de idade: {std_age:.2f}")

if std_education > std_experience and std_education > std_age:
    print("\nAnos de educação apresenta maior variabilidade.")
elif std_experience > std_education and std_experience > std_age:
    print("\nAnos de experiência apresenta maior variabilidade.")
else:
    print("\nIdade apresenta maior variabilidade.")
print("Justificativa: A variabilidade é determinada pelo maior valor de desvio-padrão entre as variáveis, indicando que os dados se afastam mais da média.")
```

q-) Determine dentre as duas variáveis: T_ANALF25M (Taxa de analfabetismo - 25 anos ou mais) ou RDPC (Renda per capita média), qual apresenta maior variabilidade. Justifique a sua resposta.

```
std_analfabetismo = df_atlas['T_ANALF25M'].std()
mean_analfabetismo = df_atlas['T_ANALF25M'].mean()
cv_analfabetismo = (std_analfabetismo / mean_analfabetismo) * 100

std_rdpc = df_atlas['RDPC'].std()
mean_rdpc = df_atlas['RDPC'].mean()
cv_rdpc = (std_rdpc / mean_rdpc) * 100

print(f"Coeficiente de Variação (CV) para T_ANALF25M: {cv_analfabetismo:.2f}%")
print(f"Coeficiente de Variação (CV) para RDPC: {cv_rdpc:.2f}%")

if cv_analfabetismo > cv_rdpc:
    print("\nA variável T_ANALF25M (Taxa de analfabetismo - 25 anos ou mais) apresenta maior variabilidade.")
else:
    print("\nA variável RDPC (Renda per capita média) apresenta maior variabilidade.")
print("Justificativa: O Coeficiente de Variação (CV) é utilizado para comparar a variabilidade entre conjuntos de dados com diferentes escalas. A variável com o maior CV apresenta maior dispersão relativa dos dados em relação à sua média.")
```
